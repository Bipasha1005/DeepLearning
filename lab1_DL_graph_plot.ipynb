```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)
y = np.maximum(0, x)

fig, ax = plt.subplots()
ax.set(title='RELU')
ax.grid()
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
y = np.maximum(0.1*x, x)

fig, ax = plt.subplots()
ax.set(title='Leaky RELU')
ax.grid()
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-100, 100, 100)
y=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

fig, ax = plt.subplots()
ax.set(title='Tanh plot')
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-10, 10, 10)
y=1/(1+np.exp(-x))

fig, ax = plt.subplots()
ax.set(title='Sigmoid Plot')
plt.plot(x,y)
plt.show()
```
