```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)

fig, ax = plt.subplots()
ax.set(title='Sine Plot')
plt.plot(x, y, color= 'red')
plt.show()
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.cos(x)

fig, ax = plt.subplots()
ax.set(title='Cosine graph')
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x) + np.cos(x)

fig, ax = plt.subplots()
ax.set(title='Sin(x)+Cos(X) Plot')
plt.plot(x, y)
plt.show()
```

```python
t = np.arange(0.0, 1, 0.01)
s = np.tan(t)

fig, ax = plt.subplots()
ax.set(xlabel='radians', ylabel='tan(x)', title='Tangent Plot')
plt.grid()
ax.set(title='Tangent graph')
plt.plot(t, s)
plt.show()
```

```python
x = np.arange(0.1, 10.0, 0.1)
y = np.log(x)

fig, ax = plt.subplots()
ax.set(title='Log(x) Plot')
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.exp(x)

fig, ax = plt.subplots()
ax.set(title='exp(x) Plot')
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.exp(-x)

fig, ax = plt.subplots()
ax.set(title='exp(-x) Plot')
plt.plot(x, y)
plt.show()
```
